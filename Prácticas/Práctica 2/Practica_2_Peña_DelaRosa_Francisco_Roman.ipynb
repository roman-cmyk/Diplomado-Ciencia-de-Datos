```python
# This function is used to reads Whatsapp '.txt' file into a list of strings
def read_file(file):
    txt_file = open(file,'r', encoding = 'utf-8')
    y = txt_file.read()
    # "splitline" method converts the chunk of string into a list of strings
    content = y.splitlines()
    return content

chat = read_file(WhatsAppChat)
```

## (d) Clean the variable "v_user", you should only have two categories in this column, the text should not contain special characters, in lowercase and without spaces.

```python
# This function is used to remove special marks/characters
def remove_punct(text):
    try:
        text=text.replace(".",' ').replace(";",' ').replace(":",' ').replace(",",' ')
        text=text.replace("(",' ').replace(")",' ').replace("|",' ').replace('"',' ')
        text=text.replace("%",' ').replace("$",' ').replace("/",' ').replace('\'',' ')
        text=text.replace("-",' ').replace("_",' ').replace("*",' ').replace('+',' ')
        text=text.replace("#",' ').replace("@",' ').replace("!",' ').replace('?',' ')
        text=text.replace("[",' ').replace("]",' ').replace("'",' ').replace('¡',' ')
        text=text.replace("í",'i').replace("é",'e').replace("á",'a').replace('ó','o').replace('ú','u')
    except:
        pass
    return text 
```

```python
# This function is used to clean 'name' and convert to lowercase
def clean_text(text):
    text=text.lower()
    text=remove_punct(text)
    return text
```

```python
# This function is used to count words and emojis in each message
import emoji
import regex

def split_count(text):
    emoji_counter = 0
    data = regex.findall(r'\X', text)
    for word in data:
        if any(char in emoji.UNICODE_EMOJI for char in word):
            emoji_counter += 1
            # Remove from the given text the emojis
            text = text.replace(word, '') 

    words_counter = len(text.split())

    return emoji_counter, words_counter
```

```python
# Create 'c_EmojisNumber' variable
df['c_EmojisCounter'] = df['t_texto'].apply(lambda row: split_count(row))
df
```
